```python
df = pd.read_csv('../data/train.csv', encoding = 'utf-8')
평균무게 = df['신고중량(KG)'].mean()
df = df[(df['신고중량(KG)'] >= 평균무게)]
corrdf = df[['신고중량(KG)','우범여부']].corr()

max_corr = corrdf["우범여부"][0]
max_wgt = 평균무게

print("max_corr : %f \nmax_wgt : %f"%(max_corr,max_wgt))

while(평균무게<df['신고중량(KG)'].max()):
    df = df[(df['신고중량(KG)'] >= 평균무게)]
    corrdf = df[['신고중량(KG)','우범여부']].corr()
    if corrdf["우범여부"][0] > max_corr:
        max_corr = corrdf["우범여부"][0]
        max_wgt = 평균무게
    평균무게 += 100

print("max_corr : %f \nmax_wgt : %f"%(max_corr,max_wgt))
```

```python
df = pd.read_csv('../data/train.csv', encoding = 'utf-8')
평균무게 = df['신고중량(KG)'].mean()
df = df[(df['신고중량(KG)'] < 평균무게)]
corrdf = df[['신고중량(KG)','우범여부']].corr()

max_corr = corrdf["우범여부"][0]
max_wgt = 평균무게

print("max_corr : %f \nmax_wgt : %f"%(max_corr,max_wgt))

while(평균무게>df['신고중량(KG)'].min()):
    df = df[(df['신고중량(KG)'] < 평균무게)]
    corrdf = df[['신고중량(KG)','우범여부']].corr()
    if (corrdf["우범여부"][0] > max_corr) or (corrdf['우범여부'][0]>max_corr * -1):
        max_corr = corrdf["우범여부"][0]
        max_wgt = 평균무게
    평균무게 -= 100

print("max_corr : %f \nmax_wgt : %f"%(max_corr,max_wgt))
```
